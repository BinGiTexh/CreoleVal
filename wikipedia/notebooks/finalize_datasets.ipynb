# Finalize datasets

```python
outputdir = "../ZS_BERT/output/01"
```

```python
finaldir = "../data/relation_extraction/final-01-09-2023"
```

```python
for lang in langs:
    file = os.path.join(outputdir, f"{lang}.json")
    csvfile = os.path.join(finaldir, f"{lang}.csv")
    jsonfile = os.path.join(finaldir, f"{lang}.json")
    with open(file) as f:
        data = json.load(f)

    header = "Sentence,ENT1,ENT2,ENT1_Qcode,ENT2_Qcode,Property\n"
    csvwriter = open(csvfile, "w+")
    jsonwriter = open(jsonfile, "w+")
    
    csvwriter.write(header)

    new_data=[]
    
    for line in data:
        tokens = line["tokens"]
        sent = " ".join(tokens)
        left = line["edgeSet"]["left"]
        right = line["edgeSet"]["right"]
        triple = line["edgeSet"]["triple"]
        ent1= triple[0]
        ent2=triple[1]
        prop = triple[2]

        ent1_str= " ".join([tokens[i] for i in left])
        ent2_str= " ".join([tokens[i] for i in right])
        line = f"{sent},{ent1_str},{ent2_str},{ent1},{ent2},{prop}\n"
        csvwriter.write(line)
        new_data.append({
            "tokens":tokens,
            "edgeSet":{
                "left":left,
                "right":right,
                "triple":triple
            }
        })
    json.dump(new_data, jsonwriter)

    jsonwriter.close()
    csvwriter.close()

        
        
```

```python
def select_json(lang):
    print(f"language {lang}")
    langdf = pd.read_csv(os.path.join(selected_dir, f"{lang}.csv"))
    sentences = langdf.Sentence.to_list()

    with open(os.path.join(results_dir, f"{lang}.json")) as f:
        data = json.load(f)
    
    lang_new = []
    triples = []

    for line in data:
        sent = " ".join(line["tokens"])
        
        t = tuple(line["edgeSet"]["triple"])
        # remove duplicate triples
        if t not in triples:     
            triples.append(t) 
            lang_new.append(line)
        else:
            print(sent)
    # randomly choose 97 samples
    lang_new = random.sample(lang_new, 97)
    print(f"len {len(lang_new)}")
    with open(os.path.join(selected_dir, f"{lang}.json"), "w+") as f:
        json.dump(lang_new, f)
```
