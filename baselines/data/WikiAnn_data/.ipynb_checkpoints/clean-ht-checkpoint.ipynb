```python
with open('ht/wikiann-ht.bio', 'r', encoding='utf8') as f:
    ht = []
    for token in f:
        ht.append(token)
```

```python
import re
```

```python
ht
```

```python
sents = {}
k = 1
for tok in ht:
    if tok != '\n':
        sents.setdefault(k, []).append(tok)
    elif tok == '\n':
        k += 1        
```

```python
sents
```

```python
regex = re.compile('[*](Afta)[*]')
```

```python
for key, value in sents.items():
    value = ''.join(value)
    if regex.match(value) is True:
        print(value)
```

```python
cleaned_sents = {}
for key, value in sents.items():
    value = '+'.join(value)
    if value not in cleaned_sents:
        cleaned_sents[value] = 1
    elif value in cleaned_sents:
        cleaned_sents[value] += 1
```

```python
cleaned_sents
```

```python
duplicates = {}
n = 0
for key, value in cleaned_sents.items():
    if value > 1:
        key = key.split('+')
        n += 1
        duplicates[n] = key
```
